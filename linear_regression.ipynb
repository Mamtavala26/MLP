```python
# Pandas
import pandas as pd

#matplotlib
import matplotlib.pyplot as plt

#linear_model
from sklearn import linear_model

# train test split
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("C:\\Users\\91976\\Desktop\\Carseats.csv")
data.head()
```

```python
plt.scatter(data['Sales'],data['Population'])
plt.show()
```

```python
# Dropping Null values
data.dropna(inplace=True)
```

```python
sliced_x = data.iloc[:,0:3].join(data.iloc[:,4:8])
sliced_x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
toBeEncode =['Sales','CompPrice' ,'Income', 'Population','Price' ,'ShelveLoc','Age']
encoder = LabelEncoder()
for col in toBeEncode:
    sliced_x[col]= encoder.fit_transform(sliced_x[col])
```

```python
sliced_y = data.iloc[:,3].values
sliced_y
```

```python
train_x, test_x , train_y, test_y = train_test_split(sliced_x.values,sliced_y,test_size=0.3)
```

```python
regression = linear_model.LinearRegression()
regression.fit(train_x,train_y)
```

```python
# Prediction
result = regression.predict(test_x)
print(f"Result: {result}")
```
